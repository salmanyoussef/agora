```python
import requests
from typing import List

def get_dataset_id_with_tag(tag: str) -> List[str]:
    """
    Fetch all dataset IDs from data.gouv.fr that match a given tag.

    Parameters
    ----------
    tag : str
        The tag to search for (e.g. "transport", "energie", "education")

    Returns
    -------
    List[str]
        A list of dataset IDs
    """

    base_url = "https://www.data.gouv.fr/api/1/datasets/"
    params = {
        "tag": tag,
        "page_size": 100  # max page size for fewer requests
    }

    dataset_ids = []
    next_page_url = base_url

    while next_page_url:
        response = requests.get(next_page_url, params=params)
        response.raise_for_status()
        payload = response.json()

        # Extract dataset IDs from this page
        for dataset in payload.get("data", []):
            dataset_id = dataset.get("id")
            if dataset_id:
                dataset_ids.append(dataset_id)

        # Pagination: API gives full URL for next page
        next_page_url = payload.get("next_page")

        # Params are only needed for the first request
        params = None

    return dataset_ids
```

```python
import requests
from typing import Any, Dict, Optional, List

def _safe_get(d: Dict[str, Any], path: List[str], default=None):
    """Safely get nested dict keys."""
    cur = d
    for key in path:
        if not isinstance(cur, dict) or key not in cur:
            return default
        cur = cur[key]
    return cur

def _fmt_bytes(n: Optional[int]) -> str:
    if not isinstance(n, (int, float)) or n < 0:
        return "N/A"
    units = ["B", "KB", "MB", "GB", "TB"]
    i = 0
    n = float(n)
    while n >= 1024 and i < len(units) - 1:
        n /= 1024
        i += 1
    return f"{n:.2f} {units[i]}"

def _print_kv(label: str, value: Any, indent: int = 0):
    prefix = " " * indent
    if value is None or value == "" or value == [] or value == {}:
        return
    print(f"{prefix}{label}: {value}")

def get_dataset_info(dataset_id: str) -> Dict[str, Any]:
    """
    Fetch and print key information about a dataset from data.gouv.fr.

    Parameters
    ----------
    dataset_id : str
        Dataset id (e.g. '5a...')

    Returns
    -------
    Dict[str, Any]
        The full dataset JSON payload (useful for downstream processing).
    """
    # Use https (recommended) even if examples show http
    url = f"https://www.data.gouv.fr/api/1/datasets/{dataset_id}/"

    r = requests.get(url, timeout=30)
    r.raise_for_status()
    ds: Dict[str, Any] = r.json()

    print("=" * 90)
    print("DATASET SUMMARY")
    print("=" * 90)

    _print_kv("ID", ds.get("id"))
    _print_kv("Title", ds.get("title"))
    _print_kv("Slug", ds.get("slug"))
    _print_kv("URI", ds.get("uri"))
    _print_kv("Page", ds.get("page"))

    print("\n--- Description ---")
    desc_short = ds.get("description_short")
    desc = ds.get("description")
    if desc_short:
        print(desc_short)
    if desc and desc != desc_short:
        print("\n" + desc)

    print("\n--- Publisher / Ownership ---")
    org = ds.get("organization") or {}
    owner = ds.get("owner") or {}
    _print_kv("Organization", org.get("name"))
    _print_kv("Organization ID", org.get("id"))
    _print_kv("Organization slug", org.get("slug"))
    _print_kv("Organization page", org.get("page"))

    owner_name = " ".join(x for x in [owner.get("first_name"), owner.get("last_name")] if x)
    _print_kv("Owner", owner_name if owner_name else owner.get("slug"))
    _print_kv("Owner page", owner.get("page"))

    print("\n--- Dates / Update cadence ---")
    _print_kv("Created at", ds.get("created_at"))
    _print_kv("Last modified", ds.get("last_modified"))
    _print_kv("Last update", ds.get("last_update"))
    _print_kv("Frequency", ds.get("frequency"))
    _print_kv("Frequency date", ds.get("frequency_date"))
    _print_kv("Archived", ds.get("archived"))
    _print_kv("Deleted", ds.get("deleted"))

    print("\n--- Access / License ---")
    _print_kv("Access type", ds.get("access_type"))
    _print_kv("Access type reason", ds.get("access_type_reason"))
    _print_kv("Access type reason category", ds.get("access_type_reason_category"))
    _print_kv("Authorization request URL", ds.get("authorization_request_url"))
    _print_kv("Private", ds.get("private"))
    _print_kv("License", ds.get("license"))

    tags = ds.get("tags") or []
    if tags:
        print("\n--- Tags ---")
        print(", ".join(tags))

    print("\n--- Spatial / Temporal coverage ---")
    temporal = ds.get("temporal_coverage") or {}
    _print_kv("Temporal start", temporal.get("start"))
    _print_kv("Temporal end", temporal.get("end"))

    spatial = ds.get("spatial") or {}
    _print_kv("Spatial granularity", spatial.get("granularity"))
    geom_type = _safe_get(spatial, ["geom", "type"])
    coords = _safe_get(spatial, ["geom", "coordinates"])
    if geom_type:
        _print_kv("Spatial geometry type", geom_type)
    if coords:
        _print_kv("Spatial coordinates (raw)", coords)

    print("\n--- Contact points ---")
    cps = ds.get("contact_points") or []
    if not cps:
        print("No contact points listed.")
    else:
        for i, cp in enumerate(cps, start=1):
            print(f"[{i}]")
            _print_kv("  Name", cp.get("name"))
            _print_kv("  Role", cp.get("role"))
            _print_kv("  Email", cp.get("email"))
            _print_kv("  Contact form", cp.get("contact_form"))
            cp_org = cp.get("organization") or {}
            if cp_org.get("name"):
                _print_kv("  Organization", cp_org.get("name"))
            cp_owner = cp.get("owner") or {}
            cp_owner_name = " ".join(x for x in [cp_owner.get("first_name"), cp_owner.get("last_name")] if x)
            if cp_owner_name or cp_owner.get("slug"):
                _print_kv("  Owner", cp_owner_name if cp_owner_name else cp_owner.get("slug"))
            print()

    print("\n" + "=" * 90)
    print("RESOURCES (MAIN + COMMUNITY)")
    print("=" * 90)

    def print_resources(resources: List[Dict[str, Any]], label: str):
        print(f"\n{label} (count={len(resources)})")
        if not resources:
            return
        for i, res in enumerate(resources, start=1):
            print("-" * 90)
            print(f"[{i}] {res.get('title') or '(no title)'}")
            _print_kv("  Resource ID", res.get("id"))
            _print_kv("  Type", res.get("type"))
            _print_kv("  Format", res.get("format"))
            _print_kv("  MIME", res.get("mime"))
            _print_kv("  Filetype", res.get("filetype"))
            _print_kv("  Filesize", _fmt_bytes(res.get("filesize")))
            _print_kv("  Created at", res.get("created_at"))
            _print_kv("  Last modified", res.get("last_modified"))
            _print_kv("  URL", res.get("url"))
            _print_kv("  Preview URL", res.get("preview_url"))

            schema = res.get("schema") or {}
            if schema:
                schema_str = " | ".join(
                    x for x in [schema.get("name"), schema.get("version"), schema.get("url")] if x
                )
                _print_kv("  Schema", schema_str)

            if res.get("description"):
                print("  Description:")
                print("   " + str(res["description"]).replace("\n", "\n   "))

    print_resources(ds.get("resources") or [], "Dataset resources")
    print_resources(ds.get("community_resources") or [], "Community resources")

    # Dataset-level schema (if present)
    ds_schema = ds.get("schema") or {}
    if ds_schema:
        print("\n" + "=" * 90)
        print("DATASET SCHEMA")
        print("=" * 90)
        schema_str = " | ".join(x for x in [ds_schema.get("name"), ds_schema.get("version"), ds_schema.get("url")] if x)
        print(schema_str if schema_str else ds_schema)

    return ds
```
